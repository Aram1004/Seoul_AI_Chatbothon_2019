```python
from bs4 import BeautifulSoup
import urllib.request as req
from selenium import webdriver
from time import sleep
import pandas as pd
from pandas import DataFrame


driver = webdriver.Chrome(r"C:\Users\dkfka\help\chromedriver.exe")
url = 'https://search.naver.com/search.naver?sm=tab_sly.hst&where=kin&query=%EB%85%B9%EB%AC%BC&oquery=%EB%85%B9%EB%AC%BC&tqi=UiOOjsprvhGssb8BFCNssssss5V-031098&acr=1'
question_list = []
for i in range(1, 100):
    link = url.format(i)
    driver.get(link)
    driver.implicitly_wait(2)

    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    comment = soup.find_all('dt',{'class':'question'})
    for review in comment:
        question_list.append(review.get_text().strip())

# for question in question_list:
#     print(question+'\n')
    
#질문들 헤더 한줄에 출력

answer_list = []
for i in range(1, 100):
    link = url.format(i)
    driver.get(link)
    driver.implicitly_wait(2)

    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    comment = soup.find_all('dd',{'class':'answer'})
    for review in comment:
        answer_list.append(review.get_text().strip())

# for answer in answer_list:
#     print(answer +'\n')

data = {'question':question_list,'answer':answer_list}

columns = ['question','answer']
df = pd.DataFrame(data, columns=columns)
df

```

```python
#질문 내용 전체 출력

from bs4 import BeautifulSoup
import urllib.request as req
from selenium import webdriver

url = 'https://m.search.naver.com/search.naver?sm=tab_hty.top&where=m_kin&query=%EB%85%B9%EB%AC%BC&oquery=%EB%85%B9%EB%AC%BC+%EB%AC%B8%EC%A0%9C%EC%A0%90&tqi=UiN2mwp0Jy0ssZ%2Fn1Vwssssssms-068629'
question_list = []
for i in range(1, 2):
    link = url.format(i)
    driver.get(link)
    driver.implicitly_wait(2)

    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    comment = soup.find_all('div', 'api_txt_lines')
    for review in comment:
        question_list.append(review.get_text().strip())

for question in question_list:
    print(question +'\n')
```
